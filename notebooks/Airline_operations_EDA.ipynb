```python
Q1= df["Ticket_Price"].quantile(0.25)
Q3= df["Ticket_Price"].quantile(0.75)
IQR = Q3-Q1
lower_limit = Q1-1.5*IQR
upper_limit= Q3+1.5*IQR
outliers = df[(df["Ticket_Price"]<lower_limit)|(df["Ticket_Price"]>upper_limit)]
```

```python
route_stats= df.groupby("Route").agg(
    avg_delay=("Delay_minutes","mean"),
    avg_price=("Ticket_Price","mean"),
    avg_fuel=("Fuel_cost_per_km","mean"),
    flights = ("Route","count")
)
```

```python
route_stats=route_stats[route_stats["flights"]>50]
```

```python
worst_routes= route_stats.sort_values(["avg_delay","avg_price","avg_fuel"],ascending=[False,True,False])
worst_routes.head(2)
```
